# Practice backtest 02

For a list of ProShares Leveraged ETFs, we check each day if the price is above or below its respective 100 day moving average.

If the price is above moving average, we will be long notional allocation of 1/Number of ETFs. That means in a perfect bull market, if all ETFs are above their respective moving averages, we will have an overall portfolio exposure of 100%, but most of the time it will be below that,

```python
%matplotlib inline

import pandas as pd

from zipline import run_algorithm
from zipline.api import order_target_percent, symbol, set_benchmark

from datetime import datetime
import pytz


import matplotlib.pyplot as plt

import pyfolio as pf

def initialize(context):

    # Which ETF to trade
    etfs = ['TQQQ', 'UPRO', 'TMF']

    # Make a list of symbols from tickers
    context.etfs = [symbol(s) for s in etfs]
    
    # MA window
    context.index_average_window = 100

    set_benchmark(symbol('SPY'))
    
def handle_data(context, data):
    # request history
    etf_hist = data.history(context.etfs, 'close', context.index_average_window, '1d')
    
    # Make an empty DF to start
    etf_analytics = pd.DataFrame()
    
    # Add column for above average
    etf_analytics['above_mean'] = etf_hist.iloc[-1] > etf_hist.mean()
    
    # Set weight for ETF to buy
    etf_analytics.loc[etf_analytics['above_mean'] == True, 'weight'] = 1 / len(context.etfs)
    
    # Set weight as zero for rest
    etf_analytics.loc[etf_analytics['above_mean'] == False, 'weight'] = 0.0
    
    # Iterate each row and place trade
    for etf, analytics in etf_analytics.iterrows():
        # Check if etf can be traded
        if data.can_trade(etf):
            # Place the trade
            order_target_percent(etf, analytics['weight'])
    
def analyze(context, perf):
    fig = plt.figure(figsize=(12, 8))
    
    # first chart
    ax = fig.add_subplot(311)
    ax.set_title("Strategy Results")
    ax.semilogy(perf['portfolio_value'], linestyle='-', label = 'Equity Curve', linewidth=3.0)
    ax.legend()
    ax.grid(True)
    
    # second chart
    ax = fig.add_subplot(312)
    ax.plot(perf['gross_leverage'], label = 'Exposure', linestyle = '-', linewidth = 1.0)
    ax.legend()
    ax.grid(True)
    
    # third chart
    ax = fig.add_subplot(313)
    ax.plot(perf['returns'], label = 'Returns', linestyle = '-.', linewidth = 1.0)
    ax.legend()
    ax.grid(True)
    
    returns, positions, transactions, leverage = pf.utils.extract_rets_pos_txn_from_zipline(perf)
    pf.create_full_tear_sheet(returns, positions = positions, transactions = transactions, round_trips = True)


# set start and end date
#start_date = datetime(2010, 2, 11, tzinfo=pytz.UTC)
#end_date = datetime(2020, 8, 15, tzinfo=pytz.UTC)
start_date = pd.to_datetime('2010-02-11', utc=True)
end_date = pd.to_datetime('2020-09-18', utc=True)


# start backtest
results = run_algorithm(
    start = start_date,
    end = end_date,
    initialize = initialize,
    analyze = analyze,
    handle_data = handle_data,
    capital_base = 100000,
    data_frequency = 'daily',
    bundle = 'etf_db_data'
)
```
